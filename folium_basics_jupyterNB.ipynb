now we need to calculate the mean of the latitude and longitude #not mandatory but some times it is useful

```python
lat_mean = air_passengers["latitude"].mean()
log_mean = air_passengers["longitude"].mean()
```

```python
#use of foilium
Busiest_AP = folium.Map(location=[lat_mean, log_mean], zoom_start=5)
```

the above writen code is just a sample which demonstrates how to use popups
know we will plot different coloured markers on the map basing on:
annual passengers above and below the mean

```python
passenger_mean_2017_2018 = air_passengers["pass_ 2017_18"].mean() #taking mean of total number of passengers(note:mean calculated by including each airport only for understanding)
```

```python
for i in range(0,len(air_passengers)):
    if air_passengers["pass_ 2017_18"][i] > passenger_mean_2017_2018:
        folium.Marker([air_passengers['latitude'][i],air_passengers['longitude'][i]],icon=folium.Icon(color='green'), popup="air passengers count = " + air_passengers['pass_ 2017_18'][i].astype(str)).add_to(Busiest_AP)
    else:
        folium.Marker([air_passengers['latitude'][i],air_passengers['longitude'][i]],icon=folium.Icon(color='orange'), popup="air passengers count = " + air_passengers['pass_ 2017_18'][i].astype(str)).add_to(Busiest_AP)
folium.LatLngPopup().add_to(Busiest_AP)
Busiest_AP
```
